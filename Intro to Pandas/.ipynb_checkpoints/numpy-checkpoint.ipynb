LINEAR ALGEBRA

```python
B = np.array([
    [6, 5],
    [4, 3],
    [2, 1]
])
```

```python
A.dot(B)
```

```python
A @ B
```

```python
B.T
```

```python
A
```

SIZE OF OBJECTS IN MEMORY

Int, floats

```python
# An integer in python is > 24bytes
sys.getsizeof(1)
```

```python
# Longs are even larger 
sys.getsizeof(10**100)
```

```python
# NumPy size is much smaller
np.dtype(int).itemsize
```

```python
# NumPy size is much smaller
np.dtype(np.int8).itemsize
```

Lists are even larger

```python
# A one-element list
sys.getsizeof(1)
```

```python
# An array of one element is bumpy
np.array([1]).nbytes
```

And performance is also important

```python
l = list(range(1000))
```

```python
a = np.arange(1000)
```

```python
%time np.sum(a ** 2)
```

```python
%time sum([x ** 2 for x in l])
```
