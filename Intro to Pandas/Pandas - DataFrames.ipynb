**Indexing, Selection and Slicing**

Individual columns in the DataFrame can be selected with regular indexing. Each column is represented as a *series*:

```python
df['Population']
```

Note that the *index* of the returned Series is the same as the DataFrame one. And its name is the name of the column. If you're working on a notebook and want to see more DataFrame-like format you can use the to_frame method:

```python
df['Population'].to_frame()
```

```python
df[['Population']]
```

Multiple columns can also be selected similarly to numpy and Series:

```python
df[['Population', 'GDP']]
```

In this case, the result is another DataFrame. Slicing works differently, it acts at row level, and can be counter intuitive

```python
df[1:3]
```

Row level selection works better with *loc* and *iloc* **which are recommended** over regular direct slicing (df[:]). *loc* selects rows matching the given index:

```python
df
```

```python
df.loc['Canada']
```

```python
df.iloc[-1]
```

```python
df.loc['Italy']
```

```python
df.loc['France': 'Italy']
```

As a second "arguement", you can pass the column(s) you'd like to select:

```python
df.loc['France': 'Italy', 'Population']
```

```python
df.loc['France': 'Italy', ['Population','GDP']]
```

*iloc* works with the (numeric) "position" of the index:

```python
df
```

```python
df.iloc[0]
```

```python
df.iloc[-1]
```

```python
df.iloc[[0, 1, -1]]
```

```python
df.iloc[1:3]
```

```python
df.iloc[1:3, 3]
```

```python
df.iloc[1:3, [0, 3]]
```

```python
df.iloc[1:3, 1:3]
```

**RECOMMENDED: Always use loc and iloc to reduce ambiguity, specially with DataFrames with numeric indexes**

<div style="height:20px; background: linear-gradient(to right, #ff7e5f, #feb47b); border-radius:2px;"></div>

We saw conditional selection applied to Series and it'll work in the same way for DataFrames. After all, a DataFrame is a collection of Series 

```python
df
```

```python
df['Population'] > 70
```

```python
df.loc[df['Population'] > 70]
```

The boolean matching is done at index level, so you can filter by row, as long ax it contains the right indexes. Column selection still works as expected:

```python
df.loc[df['Population'] > 70, 'Population']
```

```python
df.loc[df['Population'] > 70, ['Population', 'GDP']]
```

**Dropping stuff**

Opposed to the concept of selection, we have "dropping". Instead of pointing out which values you'd like to select you could point point which one you'd like to drop:

```python
df.drop('Canada')
```

```python
df.drop(['Canada', 'Japan'])
```

```python
df.drop(columns=['Population', 'HDI'])
```

```python
df.drop(['Italy', 'Canada'], axis=0)
```

```python
df.drop(['Population', 'HDI'], axis=1)
```

```python
df.drop(['Population', 'HDI'], axis=1)
```

```python
df.drop(['Population', 'HDI'], axis='columns')
```

```python
df.drop(['Canada', 'Germany'], axis='rows')
```

All these drop methods return a new DataFrame. If you'd like to modify it "in place", you can use the inplace attribute(there's an example below)

**Modifying DataFrames**

It's simple and intuitive, You can add columns, or replace values for columns without issues:

**Adding a new column**

```python
langs = pd.Series(
    ['French', 'German', 'Italian'],
    index=['France', 'Germany', 'Italy'],
    name='Language'
)
```

```python
df['Language'] = langs
```

```python
conts = pd.Series(
    ['North America', 'Europe', 'Europe', 'Europe', 'Asia', 'Europe', 'North America'],
    index=['Canada', 'France', 'Germany', 'Italy', 'Japan', 'United Kingdom', 'United States'],
    name='Yes'
)
```

```python
df['Yes'] = conts
```

```python
df
```

```python
df['Language'] = 'English'
df
```

```python
df.drop(['Yes'], axis=1)
```

**Renaming Columns**

```python
df.rename(
    columns={
        'HDI': 'Human Development Index',
        'Anual Popcorn Consumption': 'APC'
    }, index={
        'United States': 'USA',
        'United Kingdom': 'UK',
        'Argentina': 'AR'
    }
)
```

```python
df
```

```python
df.rename(index=str.upper)
```

```python
df.rename(index=lambda x: x.lower())
```

**Dropping Columns**

```python
df.drop(columns='Language', inplace=True, errors='ignore')
```

**Adding values**

```python
df = pd.concat([
    df,
    pd.Series({
        'Population': 3,
        'GDP': 5
    }, name='China')
    .to_frame()
    .T
])
```

Append returns a new DataFrame:

```python
df
```

```python
df.drop(['Yes'], inplace=True, errors='ignore', axis=1)
```

You can directly set the new index and values to the DataFrame:

```python
df.loc['China'] = pd.Series({'Population': 1_400_000_000, 'Continent': 'Asia'})
```

```python
df
```

We can use drop to just remove a row by index

```python
df.drop('China', inplace=True)
```

```python
df
```

**More radical index changes**

```python
df.reset_index()
```

```python
df.set_index('Population')
```

**Creating columns from other columns**

Altering a DataFrame often involves combining different columns into another. For example, in our Countries analysis, we could try calculate the "GDP per capita", which is just, GDP / Polulation.

```python
df[['Population','GDP']]
```

The regular pandas way of expressing that, is just dividing each series:

```python
df['GDP'] / df['Population']
```

The result of that operation is just another series that you can add to the original DataFrame:

```python
df['GDP Per Capita'] = df['GDP'] / df['Population']
```

```python
df
```
