**Pandas Series**

We'll start analyzing "The Group of seven". Which is a political formed by Canada, France, Germany, Italy, Japan, the United States. We'll start by analyzing population and for that, we'll use a pandas.Series object

```python
# In millions
g7_pop = pd.Series([35.467, 63.951, 80.940, 60.665, 127.061, 64.511, 318.523])
```

```python
g7_pop
```

Someone might not know we're representing population in millions of inhabitants. Series can have a name, to better document the purpose of the Series:

```python
g7_pop.name = 'G7 Population in millions'
```

```python
g7_pop
```

Series are pretty similar to numpy arrays:

```python
g7_pop.dtype
```

```python
g7_pop.values
```

They are actually backed by numpy arrays:

```python
type(g7_pop.values)
```

And they look like simple Python lists or Numpy Arrays. But they are actually more similar to Python dict s.

A Series has an index, that's similar to the automatic index assigned to Python's lists:

```python
g7_pop[0]
```

```python
g7_pop[1]
```

```python
g7_pop.index
```

But, in contrast to list, we can explicitly define the index

```python
g7_pop.index = [
    'Canada',
    'France',
    'Germany',
    'Italy',
    'Japan',
    'United Kingdom',
    'United States',
]
```

```python
g7_pop
```

we can say that Series look like "ordered dictionaries". We can actually create Series out of dictionaries:

```python
pd.Series({
    'Canada': 36.467,
    'France': 63.951,
    'Germany': 80.94,
    'Italy': 60.665,
    'Japan': 127.061,
    'United Kingdom': 64.511,
    'United States': 318.523
}, name= 'G7 Population in millions')
```

You can also create series out of other series, specifying indexes:

```python
pd.Series(g7_pop, index=['France', 'Germany', 'Italy', 'Spain'])
```

<font size="3">**Indexing**</font>

Indexing works similarly to lists and dictionaries, you use index of the element you're looking for:

```python
g7_pop['Canada']
```

```python
g7_pop['Japan']
```

Numeric positions can also be used, with the iloc attribute:

```python
g7_pop.iloc[0]
```

```python
g7_pop.iloc[-1]
```

Selecting multiple elements at once:

```python
g7_pop[['Italy', 'France']]
```

(The result in another Series)

```python
g7_pop.iloc[[0, 1]]
```

Slicing also works , but important, in pandas, the upper limit is also included:

```python
g7_pop['Canada': 'Italy']
```

**Conditional Selection (boolean arrays)**

The same boolean array techniques we saw applied to bumpy arrays can be used for Pandas Series:

```python
g7_pop
```

```python
g7_pop > 70
```

```python
g7_pop[g7_pop> 70]
```

```python
g7_pop.mean()
```

```python
g7_pop[g7_pop > g7_pop.mean()]
```

```python
g7_pop.std()
```

```python
g7_pop[(g7_pop > g7_pop.mean() - g7_pop.std() / 2) | (g7_pop > g7_pop.mean() + g7_pop.std() / 2)]
```

**Operations and methods**

Series also support vectorized operations and aggregation functions as Numpy

```python
g7_pop * 1000000
```

```python
g7_pop.mean()
```

```python
np.log(g7_pop)
```

```python
g7_pop['France': 'Italy'].mean()
```

**Boolean arrays**

(Work in the same way as numpy)

```python
g7_pop
```

```python
g7_pop > 80
```

```python
g7_pop[g7_pop > 80]
```

```python
g7_pop[(g7_pop > 80) | (g7_pop < 40)]
```

```python
g7_pop[(g7_pop > 80) & (g7_pop < 200)]
```
